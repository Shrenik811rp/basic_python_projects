```python
import pandas as pd
import neattext.functions as nfx
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity,linear_kernel
```

```python
data = pd.read_csv("data/udemy_courses.csv")
data.head()
```

```python
data.columns
```

```python
data['course_title']
```

```python
#clean test and remove unnecessary characters
data["new_course_title"] = data["course_title"].apply(nfx.remove_stopwords)
```

```python
data.head()
```

```python
data["new_course_title"] = data["course_title"].apply(nfx.remove_special_characters)
```

```python
data.head()
```

```python
data[["course_title","new_course_title"]]
```

```python
count_vec = CountVectorizer()

cv_mat = count_vec.fit_transform(data["new_course_title"])

print(cv_mat)
```

```python
cv_mat.todense()
```

```python
data_cv_words = pd.DataFrame(cv_mat.todense(),columns=count_vec.get_feature_names())

data_cv_words.head()
```

```python
#Cosine similarity
cosine_sim = cosine_similarity(cv_mat)

#similarity matrix
print(cosine_sim)
```

```python
#Get the course Id /index
course_index = pd.Series(data.index,index=data["course_title"]).drop_duplicates()

print(course_index)
```

```python
course_index["How To Maximize Your Profits Trading Options"]
```

```python

'''
Here we can give sample input to test which course matches with the users input

Then we will get a score for the similarity betw users and database value
'''

#User inputs a course 
ID = course_index["How To Maximize Your Profits Trading Options"]

scores = list(enumerate(cosine_sim[ID]))


'''
here below it prints out the score of how similar two items are...

So here we see that ID =4 is similar to the one we have in our dataset

so the similarity is about 99% --> (4, 0.9999999999999997)

So we have found a similar item in our DB

Which means the fourth item in the list is similar to the query user has selected

'''
print(scores)
```

```python
#sort the scores as per cosine similarity

#sort list from highest similarity to lowest similarity
sorted_scores = sorted(scores,key=lambda x:x[1],reverse=True)

#top 20 or n results
print(sorted_scores[:10])
```

```python
#Select the course indeices

#return only the index values of the recommended courses
select_course_indx = [i[0] for i in sorted_scores[:10]]

print(f"Recommended course ids: \n\n{select_course_indx}")
```

```python
#Select the course Similarity scores

#return only the index values of the recommended courses
select_course_scores = [i[1] for i in sorted_scores[:10]]

print(f"Recommended course scores: \n\n{select_course_scores}")
```

```python
final_result = data["course_title"].iloc[select_course_indx]
print(f"Final recommended list : \n\n{final_result}")
```

```python
rec_data = pd.DataFrame(final_result)

rec_data.head()
```

```python
rec_data["similarity_scores"] = select_course_scores

print(rec_data)
```
